```python
#20194111 최민규
import numpy as np
import matplotlib.pyplot as plt
from sklearn.cluster import KMeans
from sklearn import datasets
```

```python
##교재 내 코드

#X = 군집화를 수행할 데이터 포인트를 담은 NumPy 배열
X=np.array([
    [6,3],[11,15],[17,12],[24,10],[20,25],[22,30],
    [85,70],[71,81],[60,79],[56,52],[81,91],[80,81]
])

#데이터를 산점도로 시각화
plt.scatter(X[:,0],X[:,1])

#군집의 개수를 1에서 9까지의 범위로 정의한 리스트
n_clusters=range(1,10)
#n_clusters에 따라 1부터 9까지의 클러스터 개수를 갖는 KMeans 모델을 리스트로 하여 군집화 모델을 생성
kmeans=[KMeans(n_clusters=i) for i in n_clusters]
#각 클러스터 개수에 대해 KMeans 모델을 학습시킨 후 그 모델의 제곱 오차를 리스트에 추가
#(각 클러스터 개수에 대한 모델의 성능을 나타내는 지표)
score=[kmeans[i].fit(X).inertia_ for i in range(len(kmeans))]

#클러스터 개수에 따른 제곱 오차의 변화를 그래프로 표시.
plt.plot(n_clusters,score)
plt.xlabel('Numver of Clusters')
plt.ylabel('Score')
plt.title('Elbow Curve')
plt.show()
```

```python
# 붓꽃 데이터 로드
iris = datasets.load_iris()
X = iris.data

# 적절한 K 값 찾기
#SSE를 할당 할 리스트
sse = []

#클러스터 개수를 1부터 10까지 변화시키며 반복
for k in range(1, 11):
      #주어진 클러스터 개수 k로 K-평균 모델을 초기화
      kmeans = KMeans(n_clusters=k, random_state=42)
      #모델을 주어진 데이터 X에 학습
      kmeans.fit(X)
      #현재 클러스터 개수에서의 SSE를 리스트에 추가
      sse.append(kmeans.inertia_)

# 클러스터 개수에 따른 SSE의 변화를 그래프로 표시.
plt.plot(range(1, 11), sse, marker='o')
plt.xlabel('Number of Clusters (K)')
plt.ylabel('Sum of Squared Errors (SSE)')
plt.title('Elbow Method for Optimal K')
plt.show()
```
